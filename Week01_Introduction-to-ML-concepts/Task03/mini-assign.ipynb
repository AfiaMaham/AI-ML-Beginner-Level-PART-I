```python
import pandas as pd
```

```python
df = pd.read_csv("titanic/train.csv")
df.head(10)
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")

sns.countplot(x="Survived", data=df)
plt.title("Survival Count (0 = Died, 1 = Survived)")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex", data=df)
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Pclass", data=df)
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
df.groupby('Survived')[['Age', 'Fare']].mean()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x="Age", hue="Survived", kde=True, bins=30)
plt.title("Age Distribution by Survival Status")
plt.show()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df.drop('Cabin', axis=1, inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

df.isnull().sum()
```

##  EDA Summary

###  Target Variable (Survived):
- More passengers **died (0)** than **survived (1)**.

###  Gender Analysis:
- Majority of **females survived**.
- Majority of **males did not survive**.

###  Class Analysis (Pclass):
- Passengers in **1st class survived more** than those in 2nd and 3rd.
- **3rd class** had the **highest number of deaths**.

###  Age Analysis:
- Most passengers were aged **20–30**.
- Survivors were slightly **younger on average**.

###  Fare Analysis:
- Survivors paid a **higher average fare (~48)**.
- Non-survivors paid a **lower average fare (~22)**.

###  Missing Data Handling:
- Filled missing **Age** with **median**.
- Dropped **Cabin** due to many missing values.
- Filled missing **Embarked** with the most frequent value.
