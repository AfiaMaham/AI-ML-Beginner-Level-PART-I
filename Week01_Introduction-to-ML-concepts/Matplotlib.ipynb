```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([-10, 10])
y = np.array([i*i for i in x])

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = np.array(["Pakistan","Thailand", "Korea", "Japan", "China"])
y = np.array([108970, 245060, 300000, 440000, 500000])

plt.bar(x, y,color = "brown")
plt.title("Bar Graph",color= "red")

plt.show()
```

```python
x = np.array([5, 7, 8, 3, 6, 9, 2, 4, 1, 10, 15, 17, 13, 11, 16, 12])
y = np.array([45, 23, 56, 12, 45, 78, 12, 56, 34, 67, 89, 99, 44, 66, 37,19])

plt.scatter(x, y,color = "brown")
plt.title("Scatter Graph",color= "red")

plt.show()
```

```python
arr = np.random.normal(56, 10,1000)

plt.hist(arr)
plt.title("Histogram")
plt.axvline(arr.mean(), color='k', linestyle='dashed', linewidth=1)
plt.show()
```

```python
arr = np.array([np.pi/2, np.pi/3, np.pi/4, np.pi/5])
sinx = np.array([np.sin(i) for i in arr])
cosx = np.array([np.cos(i) for i in arr])
plt.suptitle("Multiple Graph")
plt.subplot(1,2,1)
plt.title("Sine Graph")
plt.plot(sinx)
plt.subplot(1,2,2)
plt.title("Cos Graph")
plt.plot(cosx)
plt.show()
```
